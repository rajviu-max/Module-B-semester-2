# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

 How could you use preattentive attributes to do that?  Suggest several possibilities.

1. You can use a different, brighter color (ex. red) to bring attention to the last 3 months and add a legend explaining what the different colors mean. 
2. You can use a thicker line and this would draw attention without using color. 
3. Make larger markers. The previous month's markers can be samller and the last three can be made bigger. This way you are taking up more visual space. 
4. You can also fade everything else out and make the last three months darker. This is similar to the color technique.

I decided to reproduce one of the ticketing graphs. 

```python
import matplotlib.pyplot as plt
```

```python
# Values read from the chart (9 data points each)
received  = [165, 240, 148, 130, 200, 168, 142, 152, 178]
processed = [190, 232, 145, 122, 158, 163, 103, 148, 133]

x = np.arange(len(received))

fig, ax = plt.subplots(figsize=(9, 4))
fig.patch.set_facecolor('white')
ax.set_facecolor('white')

ax.plot(x, received,  color='blue', linewidth=2.2, label='Received')
ax.plot(x, processed, color='red', linewidth=2.2, label='Processed')

ax.set_ylim(100, 300)
ax.set_yticks([100, 150, 200, 250, 300])
ax.set_xticks([])                         


ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.yaxis.grid(False)
ax.tick_params(axis='y', length=0, labelsize=10)

plt.tight_layout()
plt.show()
```
